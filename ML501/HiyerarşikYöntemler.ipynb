```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
warnings.filterwarnings("ignore",category=FutureWarning)
```

```python
df=pd.read_csv("USArrests.csv",index_col=0)
```

```python
df.head()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
hc_complete=linkage(df,"complete")
hc_average=linkage(df,"average")
```

```python
plt.figure(figsize=(10,5))
plt.title("Hiyerarşik k. Dendrogram")
plt.xlabel("Gözlem birimleri")
plt.ylabel("Hesaplanan uzaklıklar")
dendrogram(hc_complete,leaf_font_size=10);

plt.figure(figsize=(10,5))
plt.title("Hiyerarşik k. Dendrogram Average")
plt.xlabel("Gözlem birimleri")
plt.ylabel("Hesaplanan uzaklıklar")
dendrogram(hc_average,leaf_font_size=10);
```

```python
# 4 kümeye göre her kümede bulunan gözlem sayısı toplamları ile özelleşmiş görünüme ulaşabiliriz.
plt.figure(figsize=(10,5))
plt.title("Hiyerarşik k. Dendrogram")
plt.xlabel("Gözlem birimleri")
plt.ylabel("Hesaplanan uzaklıklar")
dendrogram(hc_complete,leaf_font_size=10,truncate_mode="lastp",p=4,show_contracted=True);
```
