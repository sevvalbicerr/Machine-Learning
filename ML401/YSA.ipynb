```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
df=pd.read_csv("diabetes.csv")
df.head()
```

```python
y=df["Outcome"]
X=df.drop(["Outcome"],axis=1)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.30)
```

```python
from sklearn.preprocessing import StandardScaler


scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)

scaler.fit(X_test)
X_test=scaler.transform(X_test)
```

```python
mlpc=MLPClassifier().fit(X_train,y_train)
```

```python
mlpc.coefs_
```

```python
?mlpc
```

```python
y_pred=mlpc.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

### Model tuning

```python
mlpc_model=MLPClassifier(activation="logistic")
```

```python
mlpc_params={"alpha":[0.1,0.01,0.03,0.001,1,5,2],
"hidden_layer_sizes":[(10,10),(100,100,100),(100,100)]}
```

```python
mlpc_cv_model=GridSearchCV(mlpc_model,mlpc_params,cv=10,n_jobs=-1).fit(X_train,y_train)
```

```python
mlpc_cv_model.best_params_
```

```python
mlpc_tuned=MLPClassifier(alpha=0.001,hidden_layer_sizes=(100,100)).fit(X_train,y_train)
```

```python
y_pred=mlpc_tuned.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
